```python
total_market_index["1980":].plot(figsize=(12,6))
total_market_index["1980":].rolling(window=36).mean().plot(figsize=(12,6))
```

```python
tmi_tr35mrets = total_market_return.rolling(window=36).aggregate(erk.annualized_rets, periods_per_year=12)
tmi_tr35mrets.plot(figsize=(12,6), label="Tr36 mo Return", legend=True)
total_market_return.plot(label="Return", legend=True)
```

### Rolling Correlation - along with MultiIndexes and .groupby

```python
ts_corr = ind_return.rolling(window=36).corr()
```

```python
ts_corr.tail()
```

```python
ts_corr.index.names = ['date', 'industry']
```

```python
int_tr36corr = ts_corr.groupby(level='date').apply(lambda cormat: cormat.values.mean())
```

```python
int_tr36corr.plot()
```

```python
tmi_tr35mrets["2007":].plot(label="Tr36 Mo Rets", figsize=(12,6), legend=True)
int_tr36corr["2007":].plot(label="Tr36 Mo Corr", figsize=(12,6), legend=True, secondary_y= True)
```
